```python
df3['gen_processed'] = df3['gen'].map({'male':1,'female':0})
```

```python
df3['gen_processed_lambda'] = df3['gen'].apply(lambda x: 1 if x=='male' else 0 )
```

```python
def encode(x):
    if x=='male':
        return 1
    else:
        return 0
```
