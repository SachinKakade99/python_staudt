```python
import pandas as pd
```

```python
df = pd.read_csv("HR_comma_sep.csv")
```

```python
df
```

```python
df.groupby('left').count()
```

```python
df['left'].value_counts()
```

```python
df1 = pd.get_dummies(df.salary)
```

```python
df1
```

```python
df_process = pd.concat([df,df1],axis=1)
```

```python
df_process
```

```python
df2 = pd.get_dummies(df['Department'])
```

```python
df_final = pd.concat([df_process,df2],axis=1).drop(columns=['Department','salary'],axis=1)
```

```python
df_final
```

```python
y = df_final['left']
```

```python
y
```

```python
X = df_final.drop('left',axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=True)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
model = LR.fit(X_train,y_train)
```

```python
model
```

```python
y_pred = model.predict(X_test)
```

```python
len(y_pred)
```

```python
len(y_test)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
f1_score(y_pred,y_test)
```
