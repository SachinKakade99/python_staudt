```python
import re
def test(x):
    obj = re.findall(r'\d+',x)
    if len(obj)>1:
        return (int(obj[0])+int(obj[1]))/2
    return obj[0]
```

```python
df['total_sqft_new'] = df['total_sqft'].apply(test)
```
